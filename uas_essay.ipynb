```python
# a function to filter the data item with grade more than 80
def higherThan80(data):
    result = []
    
    for item in data:
        if item.get("grade") > 80:
            result.append(item)
            
    return result
```

```python
# a function to display the data
def printData(data):
    for item in data:
        print("=====================")
        print("Student Number: ", item.get("student number"))
        print("Name: ", item.get("name"))
        print("Subject: ", item.get("subject"))
        print("Grade: ", item.get("grade"))
        print("=====================")
        print("\n")
```

```python
print("Data dengan nilai lebih dari 80\n")
printData(higherThan80(data))
```

```python
# import libray pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# declare pandas DataFrame from the dictionary data
dataFrame = pd.DataFrame(data)

# student number is unique, so I decided to make it the index
dataFrame = dataFrame.set_index("student number")
dataFrame
```

```python
# adding a new column called `description`
# the values is conditional based on the grade
# if the grade is lower than 70 than the description is FAIL, otherwise it PASS
dataFrame["description"] = ["FAIL" if grade < 70 else "PASS" for grade in dataFrame["grade"]]
dataFrame
```

```python
# make a grouped data frame, grouped by student number and name, 
# inside it has a column named grade which is the mean of the student's grade
groupedDataFrame = dataFrame.groupby(["student number", "name"])["grade"].mean().astype(int)
groupedDataFrame
```

```python
# make a bar chart
axis = groupedDataFrame.plot(kind="bar")

# set the y axis min to 0, and max to 100
axis.set_ylim([0, 100])

# set the x axis ticks labels to the student name
axis.set_xticklabels([x.get_text().split(',')[1][0:-1] for x in axis.get_xticklabels()])

# set the axis labels
plt.xlabel("Name")
plt.ylabel("Grade")

# set the chart title
plt.title("Mean of Student Grades Values")

# show or render the chart
plt.show()
```
